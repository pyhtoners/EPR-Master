```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df1 is your loaded dataset

# Classify each country as either Global North or Global South
def categorize_region(country):
    # List of countries in the Global North
    global_north_countries = [
    "United States of America", "Canada", "United Kingdom", "Ireland", 
    "Netherlands", "Belgium", "Luxembourg", "France", "Switzerland", 
    "Spain", "Portugal", "Germany Federal Republic", "Germany Democratic Republic", 
    "Poland", "Austria", "Hungary", "Czechoslovakia", "Czech Republic", 
    "Slovakia", "Italy/Sardinia", "Malta", "Serbia", "Montenegro", 
    "Macedonia (FYROM/North Macedonia)", "Croatia", "Yugoslavia", "Bosnia-Herzegovina", 
    "Kosovo", "Slovenia", "Greece", "Cyprus", "Bulgaria", "Moldova", 
    "Rumania", "Russia (Soviet Union)", "Estonia", "Latvia", "Lithuania", 
    "Ukraine", "Belarus (Byelorussia)", "Armenia", "Georgia", "Azerbaijan", 
    "Finland", "Sweden", "Norway", "Denmark", "Iceland", "Australia", 
    "New Zealand", "Japan"
]
    return 'Global North' if country in global_north_countries else 'Global South'

# Apply the classification
df1['Region'] = df1['countryname'].apply(categorize_region)

# Function to find the most common status for each group
def most_common_status(group):
    return group['statusname'].value_counts().idxmax()

# Group by Region and Year, and find the most common status name
most_common_status_by_year = df1.groupby(['Region', 'year']).apply(most_common_status).reset_index(name='Most Common Status')

# Separate data for Global North and South
north_data = most_common_status_by_year[most_common_status_by_year['Region'] == 'Global North']
south_data = most_common_status_by_year[most_common_status_by_year['Region'] == 'Global South']

# Visualization for Global North
plt.figure(figsize=(15, 5))
sns.lineplot(x='year', y='Most Common Status', data=north_data)
plt.title('Most Common Status by Year in Global North')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Visualization for Global South
plt.figure(figsize=(15, 5))
sns.lineplot(x='year', y='Most Common Status', data=south_data)
plt.title('Most Common Status by Year in Global South')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# List of countries provided
countries_list = [
    'Trinidad and Tobago', 'El Salvador', 'Peru', 'Brazil', 'Paraguay',
    'Belgium', 'Spain', 'Serbia', 'Macedonia', 'Bosnia and Herzegovina',
    'Russia', 'Lithuania', 'Azerbaijan', 'Guinea-Bissau', 'Tanzania',
    'Burundi', 'Angola', 'Mozambique', 'Zimbabwe', 'Malawi', 'Namibia',
    'Botswana', 'Madagascar', 'Algeria', 'South Sudan', 'Turkey', 'Iraq',
    'Israel', 'Yemen', 'China', 'Taiwan', 'Pakistan', 'Bangladesh',
    'Myanmar', 'Sri Lanka', 'Malaysia', 'Australia'
]

# Define Global North and South lists
global_north = [
    'Belgium', 'Spain', 'Serbia', 'Macedonia', 'Bosnia and Herzegovina',
    'Russia', 'Lithuania', 'Azerbaijan', 'Australia', 'Israel'
]

# All other countries will be classified as Global South
global_south = [country for country in countries_list if country not in global_north]

# Function to categorize each country
def categorize_by_region(country):
    if country in global_north:
        return 'Global North'
    elif country in global_south:
        return 'Global South'
    else:
        return 'Other'  # For countries not in the provided list

# Assuming 'df' is your DataFrame
# Apply the categorization to the 'Country' column
df2['Region'] = df2['country'].apply(categorize_by_region)

# Now you can perform analyses based on the 'Region' categorization
# For example, to plot the number of unique organizations by region for each year:

# Group by year and region, and count unique organizations
orgs_by_year_region = df2.groupby(['year', 'Region'])['orgid'].nunique().reset_index()

# Pivot the data for plotting
orgs_by_year_region_pivot = orgs_by_year_region.pivot(index='year', columns='Region', values='orgid')

# Plot
orgs_by_year_region_pivot.plot(kind='bar', stacked=True, figsize=(15, 7))
plt.title('Number of Unique Organizations by Region and Year')
plt.xlabel('Year')
plt.ylabel('Number of Unique Organizations')
plt.show()
```
